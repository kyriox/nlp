### Condicionales

Python utiliza *if*, *elif* (no else if), y *else* Utiliza dos puntos(:) para indicar niveles de anidamiento.

```python
a=-1
if a < 0:
    print('menor')
    if a%2==0:
        print("par")
elif a == 0:
    print('igual')
elif a==1:
    print("es 1")
else:
    print('mayor')
    print("algo")
```

Repetir una operacin, mientras una condicin de cumpla:

```python
num_moons = 3
while num_moons > 0:
    print(num_moons)
    num_moons -= 1
```

como interrumpir un ciclo

```python
num_moons=5
while True: # Parece un ciclo infinito...
    print(num_moons)
    num_moons -= 1
    if num_moons <= 0:
        break
```

Se puede saltar la ejecucin de una sola iteracin mediante *continue*

```python
num_moons = 5
while num_moons > 0:
    print('top:', num_moons)
    num_moons -= 1
    if (num_moons % 2) == 0:
        continue
    print('...bottom:', num_moons)
```

## Listas

Una lista es una secuencia mutable de objetos.
  * Secuencia significa que puede ser indexada (Los ndices comienzan en 0 como en C, Java y C#)
  * Mutable significa que puede ser modificada sin tener que crear una copia.
  * Puede contener cualquier tipo de datos
  * Es semejante a un array o un vector, que se redimensiona automticamente
  * La funcin (built-in) *len* regresa la longitud de una cadena

Se crean listas poniendo los valores entreparntesis cuadrados
* La lista vaca se escribe [ ]

```python
gases = ['He', 'Ne', 'Ar', 'Kr']
print(len(gases)) # Longitud de la lista
print(gases[0]) # primer elemento de la lista
print(gases)
```

### Divisin (slicing)

Lista [inicio:fin] toma una sublista. Crea una nueva lista que contiene los elementos desde inicio hasta fin (no incluye el caracter en el ndice fin)

```python
print("Elementos 2 y 3:", gases[2:4])
print("Elementos del 0 al 1", gases[:2])
print("Elementos del 4 hasta el final", gases[4:])
```

Python siempre verifica los limites cuando se accede por ndice a un solo elemento. Pero trunca cuando se toma un intervalo

```python
print(gases[1:20]) # trunca
print(gases[20])  # verifica (error)
```

Se pueden utilizar ndices negativos! 

```python
print(gases[1:-1]) # El ltimo elemento
print(gases[-2]) # El penltimo elemento
```

### Compresin de Listas 

Python soporta un concepto llamado *list comprehensions*. Este proceso puede ser utlizado para construir listas de forma simple, justo como lo hacen los matemticos.

+ $S = \{x^2 : x \in \{0 ... 9\}\}$
+ $V = (1, 2, 4, 8,\dots, 2^{12})$
+ $M = \{x | x \in S \land x$ es par $\}$

Como lo hacemos en Python?

```python
S=[i**2 for i in range(10)]
V=[2**i for i in range(10)]
M=[i for i in S if i%2==0]
print(S)
print(V)
print(M)
```

Como ejemplo implementemos la raz  cuadrada utilizando el mtodo de biseccin

```python
def mysqrt(n):
    xi,xm,xf=0,n/2.0,n
    while (abs(xm*xm-n))>0.0001:
        if xm*xm > n:
            xf=xm
        else:
            xi=xm
        xm=(xi+xf)/2
    return xm  
```
